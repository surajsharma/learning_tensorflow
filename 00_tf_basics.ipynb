### Creating random tensors / shuffle

```python
r1 = tf.random.Generator.from_seed(42)
r1 = r1.normal(shape=(3,2))


r2 = tf.random.Generator.from_seed(44)
r2 = r2.normal(shape=(3,2))

r1, r2, r1 == r2
```

```python
### Shuffle a tensor when inherent order doesn't matter
not_shuffled = tf.constant([[1,2],[3,4],[5,6], [7,8]])
tf.random.shuffle(not_shuffled)
```

```python
not_shuffled
```

```python
tf.random.shuffle(not_shuffled, seed=42) #op level random seed
```

```python
tf.random.set_seed(42) #global random seed
tf.random.shuffle(not_shuffled, seed=42) #op level random seed
```

```python
tf.ones([4,4]) #ones
```

```python
tf.zeros([3,3]) #zeros
```

 ### Tensor Attributes
 * Shape `tensor.shape`
 * Rank `tensor.ndim`
 * Axis/Dimension `tensor[0], tensor[:1]`
 * size `tf.size`

```python
# Create a rank for tensor (4 Dimensions)

rank_4_tensor= tf.zeros([2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor[1]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor

print("Datatype of every element: ", rank_4_tensor.dtype)
print("Number of dimensions (rank): ", rank_4_tensor.ndim)
print("Shape of tensor ", rank_4_tensor.shape)
print("Elements along the 0 axis: ", rank_4_tensor.shape[0])
print("Elements along the last axis: ", rank_4_tensor.shape[-1])
print("Total number of elements in our tensor: ", tf.size(rank_4_tensor))
print("Total number of elements in our tensor (Numpy): ", tf.size(rank_4_tensor).numpy())
```

### Indexing tensors

Tensors can be indexed like python lists

```python
# get first two elements of each dimension
rank_4_tensor[:2,:2,:2,:2]
```

```python
# get the first element from each dimension except the final one
rank_4_tensor[:1,:1,:1]
```

```python
rank_2_tensor = tf.constant([[2,1],[3,4]])
rank_2_tensor.shape, rank_2_tensor.ndim
```

```python
# get the last item of each of our rank 12 tensor
rank_2_tensor[:,-1]
```

```python
rank_3_tensor = rank_2_tensor[...,tf.newaxis]
rank_3_tensor
```

```python
#alternative to tf.newaxis
tf.expand_dims(rank_2_tensor, axis=-1)
```

```python
tf.expand_dims(rank_2_tensor, axis=1)
```

```python
rank_2_tensor
```
